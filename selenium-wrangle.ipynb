```python
import pandas as pd
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver

# Set up the Selenium webdriver
driver = webdriver.Chrome()
driver.get('https://www.mshp.dps.missouri.gov/HP68/SearchAction')

# Find the 'Crash County' dropdown element and select 'Lawrence'
dropdownbox = driver.find_elements(by=By.TAG_NAME, value='Option')

# click option 55
dropdownbox[55].click()

button = driver.find_elements(by=By.CLASS_NAME, value='commandExButton')

button[0].click()

elements_with_class_print = driver.find_elements(by=By.CLASS_NAME, value='infoCellPrint')
elements_with_class_2print = driver.find_elements(by=By.CLASS_NAME, value='infoCell2Print')

all_elements = elements_with_class_print + elements_with_class_2print

# Create empty DataFrames to store the scraped data
crash_df = pd.DataFrame()
vehicle_df = pd.DataFrame()
injury_df = pd.DataFrame()
misc_df = pd.DataFrame()

for i in range(len(all_elements)):
    # Re-locate the elements on the page
    elements_with_class_print = driver.find_elements(by=By.CLASS_NAME, value='infoCellPrint')
    elements_with_class_2print = driver.find_elements(by=By.CLASS_NAME, value='infoCell2Print')
    all_elements = elements_with_class_print + elements_with_class_2print

    # click down the list
    all_elements[i].click()

    # Wait for the new page to load
    time.sleep(5)

    # Extract data from the search results page using pd.read_html()
    dfs = pd.read_html(driver.page_source)

    # Select the tables on the page
    Crash_Info = dfs[0]
    Vehicle_Info = dfs[1]
    Injury_Info = dfs[2]
    Misc = dfs[3]

    # Append the tables to their respective DataFrames
    crash_df = crash_df.append(Crash_Info)
    vehicle_df = vehicle_df.append(Vehicle_Info)
    injury_df = injury_df.append(Injury_Info)
    misc_df = misc_df.append(Misc)

    # Go back to the previous page
    driver.back()

    # Wait for the previous page to load
    #time.sleep(5)
    
# Close the driver
driver.quit()

# Print the DataFrames
print(crash_df)
print(vehicle_df)
print(injury_df)
print(misc_df)
```
